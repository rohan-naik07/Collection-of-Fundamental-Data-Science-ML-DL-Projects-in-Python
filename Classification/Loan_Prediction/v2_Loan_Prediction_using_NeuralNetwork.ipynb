```python
#Inference for below function
#1. if dependent are 3+ or 2 or 1 => assign Male
#2. if income is greater than 20000 => assign Male
```

```python
def assign_gender(cols):
  gender=cols[0]
  income=cols[1]
  depend = cols[2]
  marriage = cols[3]

  if pd.isnull(gender):

    if income > 16000:
      gender='M'
      return gender

    else:
      if pd.notnull(depend):
        if depend == '3+':
          gender='M'
          return gender

        elif depend=='2':
          gender='M'
          return gender

        else:
          if depend=='1':
            gender='M'
            return gender

          elif marriage == 'Yes':
            gender='M'
            return gender
          
          else:
            gender='F'
            return gender

      else:
        if marriage=='Yes':
           gender='M'
           return gender

        else:
          gender='F'
          return gender

  else:
    return gender
```

```python
df_train['Gender'] = df_train[['Gender','ApplicantIncome','Dependents','Married']].apply(assign_gender,axis=1)
```

```python
df_test['Gender'] = df_test[['Gender','ApplicantIncome','Dependents','Married']].apply(assign_gender,axis=1)
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
x = tf.keras.utils.normalize(x,axis=1)
x_tes = tf.keras.utils.normalize(x_tes,axis=1)
```

```python
model= tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(128,input_dim=10,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(1, activation=tf.nn.sigmoid))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(x,y,epochs=2000)
```
